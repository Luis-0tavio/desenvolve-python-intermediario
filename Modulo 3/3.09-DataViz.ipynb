#### Q2.
Sobre os dados de copa do mundo, qual a distribuição de público presente nos jogos? Isso pode ser respondido com um histograma com os dados da coluna `comparecimento`.  

Lembre-se que alguns jogos estavam com público 0 incorretamente, que tal remover essas ocorrências para não atrapalhar sua visualzação?

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
#### Solução com matplotlib
# Filtrando os dados
df_q2 = wc[wc['comparecimento'] > 0]

plt.figure(figsize=(10, 5))
plt.hist(df_q2['comparecimento'], bins=40, color='skyblue', edgecolor='black')
plt.title('Distribuição de Público nos Jogos da Copa do Mundo')
plt.xlabel('Público (Comparecimento)')
plt.ylabel('Frequência (Qtd. de Jogos)')
plt.show()
````

````python
#### solução com seaborn
plt.figure(figsize=(10, 5))
sns.histplot(data=df_q2, x='comparecimento', bins=40, kde=True, color='teal')
plt.title('Distribuição de Público nos Jogos da Copa do Mundo')
plt.xlabel('Público (Comparecimento)')
plt.ylabel('Frequência (Qtd. de Jogos)')
plt.show()
````

````python
#### solução com plotly
fig = px.histogram(df_q2, x='comparecimento', nbins=40,
                   title='Distribuição de Público nos Jogos da Copa do Mundo',
                   labels={'comparecimento': 'Público (Comparecimento)'},
                   color_discrete_sequence=['#636EFA'])
fig.update_layout(yaxis_title='Frequência (Qtd. de Jogos)')
fig.show()
````

#### Q3.

Apresente um gráfico de dispersão (scatter) dos atributos `gols_1` e `gols_2`. Isso representa a relação entre gols feitos e gols tomados por jogo. Há alguma relação interessante entre esses atributos?

Para facilitar a visualização dos dados (já que tem muitos placares repetidos), aplique uma leve distorção aos dados para que cada ponto esteja deslocado aleatoriamente de seu valor original. Código apresentado a seguir
```python
gols = wc[['gols_1', 'gols_2']] * np.random.random((len(wc),2))
```

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
#### solução com matplotlib
gols_distorcidos = wc[['gols_1', 'gols_2']] * np.random.random((len(wc), 2))

# Gráfico
plt.figure(figsize=(8, 6))
plt.scatter(gols_distorcidos['gols_1'], gols_distorcidos['gols_2'], alpha=0.5, color='purple')
plt.title('Relação entre Gols do Time 1 e Time 2 (Com Distorção)')
plt.xlabel('Gols Time 1 (Distorcido)')
plt.ylabel('Gols Time 2 (Distorcido)')
plt.show()
````

````python
#### solução com seaborn
jitter = (np.random.rand(len(wc), 2) - 0.5) * 0.8
gols_jitter = wc[['gols_1', 'gols_2']].astype(float) + jitter

plt.figure(figsize=(8, 6))
sns.scatterplot(x=gols_jitter.iloc[:, 0], y=gols_jitter.iloc[:, 1], alpha=0.4)
plt.title('Relação de Gols: Time 1 vs Time 2 (Com Jitter)')
plt.xlabel('Gols Time 1')
plt.ylabel('Gols Time 2')
plt.show()
````

````python
#### solução com plotly
fig = px.scatter(x=gols_distorcidos['gols_1'], y=gols_distorcidos['gols_2'],
                 title='Relação entre Gols do Time 1 e Time 2 (Com Distorção)',
                 labels={'x': 'Gols Time 1 (Distorcido)', 'y': 'Gols Time 2 (Distorcido)'},
                 opacity=0.6)
fig.show()
````

#### Q4.

Apresente um gráfico de barras com o top 10 países que mais participaram de copas do mundo, onde no eixo x devem estar o nome dos países e no eixo y a contagem de participações. Você deve separar a contagem de participações em copas femininas e masculinas, empilhando as barras de cada informação.

No exemplo de barras empilhadas da galeria do matplotlib, imagine que a parte azul são as participações do país em copas masculinas, e em laranja as participações femininas:
* https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos:
* título
* rótulos de dimensão.
* legenda

````python
#### solução com matplotlib
# Preparação dos dados: Concatenar times 1 e 2 para contar todas as participações
part_1 = wc[['time_1', 'copa']].rename(columns={'time_1': 'time'})
part_2 = wc[['time_2', 'copa']].rename(columns={'time_2': 'time'})
participacoes = pd.concat([part_1, part_2])

# Descobrir o Top 10 que mais jogaram
top10_times = participacoes['time'].value_counts().head(10).index

# Filtrar o dataset apenas com o top 10 e criar a tabela cruzada
df_top10 = participacoes[participacoes['time'].isin(top10_times)]
df_plot = pd.crosstab(df_top10['time'], df_top10['copa'])

# Ordenar pelo total de participações para o gráfico ficar mais bonito
df_plot['total'] = df_plot.sum(axis=1)
df_plot = df_plot.sort_values('total', ascending=False).drop('total', axis=1)

# Gráfico 
df_plot.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title('Top 10 Países com Mais Partidas em Copas do Mundo')
plt.xlabel('País')
plt.ylabel('Quantidade de Partidas')
plt.legend(title='Categoria da Copa')
plt.xticks(rotation=45)
plt.show()
````

````python
#### solução com seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df_top10, x='time', hue='copa', multiple='stack', shrink=0.8, palette='Set2')
plt.title('Top 10 Países com Mais Partidas em Copas do Mundo')
plt.xlabel('País')
plt.ylabel('Quantidade de Partidas')
plt.xticks(rotation=45)
plt.show()
````

````python
#### solução com plotly
# 1. Primeiro, precisamos agrupar os dados para ter a contagem exata por país e por tipo de copa
df_plotly_bar = df_top10.groupby(['time', 'copa']).size().reset_index(name='contagem')

# 2. Criar a ordem dos países do maior para o menor (baseado no total de jogos)
ordem_paises = df_top10['time'].value_counts().index

# 3. Gráfico de Barras com Plotly (px.bar)
fig = px.bar(df_plotly_bar, 
             x='time', 
             y='contagem', 
             color='copa',
             title='Top 10 Países com Mais Partidas em Copas do Mundo',
             labels={'time': 'País', 'contagem': 'Quantidade de Partidas', 'copa': 'Categoria da Copa'},
             category_orders={'time': ordem_paises}) # Aplica a ordenação

fig.show()
````

#### Q5.

Vamos fazer um compilado com as estatísticas históricas de copas do mundo!

Com a biblioteca de sua preferência você deve criar 4 subplots organizados em um grid de 2 linhas e 2 colunas. Eles devem conter os seguintes gráficos:
* Linha 1, coluna 1: Gráfico de barras com a quantidade de jogos que aconteceram por ano
* Linha 1, coluna 2: Gráfico de área (referências a seguir) com o total de gols por ano, separando as informações de `gols_1` e `gols_2` para distinguir gols em casa e do time visitante.
* Linha 2, coluna 1: Gráfico de área com o total de cartões por ano, separando as informações de cartões amarelos e cartões vermelhos, mas agregando cartões do time 1 ou time 2. Ou seja, uma área com `cartao_amarelo_1 + cartao_amarelo_2` e outra área com `cartao_vermelho_1 + cartao_vermelho_2`.
* Linha 2, coluna 2: Gráfico de barras com o total de gols contra por ano, somando `gols_contra_1` e `gols_contra_2`.

Referências sobre gráfico de área
* Matplotlib: https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html#sphx-glr-gallery-lines-bars-and-markers-stackplot-demo-py
* Pandas + Matplotlib: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.area.html
* Seaborn: https://seaborn.pydata.org/generated/seaborn.objects.Area.html
* Plotly: https://plotly.com/python/filled-area-plots/

````python
ef contar_eventos_seguro(series):
    # Transforma strings em listas e conta, ou retorna 0 se for NaN
    return series.str.split(',').str.len().fillna(0)

wc['total_amarelos'] = contar_eventos_seguro(wc['cartao_amarelo_1']) + contar_eventos_seguro(wc['cartao_amarelo_2'])
wc['total_vermelhos'] = contar_eventos_seguro(wc['cartao_vermelho_1']) + contar_eventos_seguro(wc['cartao_vermelho_2'])
wc['total_gols_contra'] = contar_eventos_seguro(wc['gols_1_contra']) + contar_eventos_seguro(wc['gols_2_contra'])

# Agrupamentos
stats_ano = wc.groupby('ano').agg({
    'gols_1': 'sum',
    'gols_2': 'sum',
    'total_amarelos': 'sum',
    'total_vermelhos': 'sum',
    'total_gols_contra': 'sum'
})
jogos_ano = wc.groupby('ano').size()

# Plotagem do Grid 2x2 (Matplotlib)
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# 1. Jogos por Ano
axs[0, 0].bar(jogos_ano.index, jogos_ano.values, color='skyblue')
axs[0, 0].set_title('Partidas por Ano')

# 2. Gols (Área)
axs[0, 1].stackplot(stats_ano.index, stats_ano['gols_1'], stats_ano['gols_2'], 
                    labels=['Casa', 'Visitante'], colors=['#4CAF50', '#8BC34A'])
axs[0, 1].legend(loc='upper left')
axs[0, 1].set_title('Total de Gols (Stack)')

# 3. Cartões (Área)
axs[1, 0].stackplot(stats_ano.index, stats_ano['total_amarelos'], stats_ano['total_vermelhos'],
                    labels=['Amarelos', 'Vermelhos'], colors=['#FFEB3B', '#F44336'])
axs[1, 0].legend(loc='upper left')
axs[1, 0].set_title('Evolução de Cartões')

# 4. Gols Contra (Barras)
axs[1, 1].bar(stats_ano.index, stats_ano['total_gols_contra'], color='grey')
axs[1, 1].set_title('Gols Contra por Ano')

plt.tight_layout()
plt.show()
````
